```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv(r'data\train.csv')
df2 = pd.read_csv(r'data\test.csv')
print(df1.shape)
df1.head()
```

```python
print(df2.shape)
df2.head()
```

```python
data = pd.concat([df1,df2])
data
```

```python
data=data.dropna(subset=['Survived'])
```

```python
data['Has_Cabin']=data['Cabin'].isnull()
```

```python
import re 
def split_it(name):
    match = re.search("(Mr|Mrs|Master|Miss)", name)
    if match:
        return match.group()

data["Titles"] = data['Name'].apply(split_it)
data
```

```python
data=data.fillna(data.mean())
```

```python
data['Categorical_age'] = pd.qcut(data['Age'],3,labels=['Child','Adult','Senior-citizen'])

data['Categorical_fare'] = pd.qcut(data['Fare'],3,labels=['Low','Medium','High'])
```

```python
data
```

```python
data=data.drop(['PassengerId','Cabin','Age','Fare','SibSp','Parch'],axis=1)
data
```

```python
data=pd.get_dummies(data,drop_first=True)
```

```python
x=data.drop(["Survived"],axis=1)
y=data['Survived']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x,y, test_size = 0.3,random_state = 3)
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
score = classifier.score(X_test, y_test)
print(score)
```
